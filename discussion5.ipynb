## 1) Define null and alternative hypotheses

### Null Hypothesis

We define the null hypothesis to be the case that nothing is happening. Generally, this means that features or distributions are the same, or observations are from the same distribution. Whatever the "boring" answer is.
The hypothesis must be exact and measurable in order to test if it is true or not.

We define the alternative hypothesis to be a case where the null hypothesis is not true. This can be either directional or non-directional. We call it directional when we only care that the null hypothesis is false in one direction. We call it non-directional when we care about the null hypothesis being false in both directions.

## 2) Define test statistic

We have seen a number of test statistics, and there are plenty more out there we haven't even talked about. However, you can boil test statistic selection down to three criteria:

- Does the test statistic relate to the question you're asking? (This relies a lot on context)
- Does the test statistic fit the kind of test you're performing? (directional v.s. non-directional)
- Does the test statistic fit the kind of data you're comparing? (Numerical v.s. categorical)

Let's review some common test statistics from this course and see how they fit these criteria.

#### Difference in means

- Considers magnitude of observation feature values
- Directional
- Numerical

#### Difference in proportions

- Considers count of observations
- Directional
- Categorical

#### Absolute difference in means

- Considers magnitude of observation feature values
- Non-directional
- Numerical

#### Absolute difference in proportions

- Considers count of observations
- Non-directional
- Categorical

#### Total Variation Distance (TVD)

- Compares similarity of two distributions
- Non-directional
- Categorical

#### Kolmogorov–Smirnov Statistic (K-S stat)

- Compares similarity of two distributions
- Non-directional
- Numerical

## 4) Hypothesis tests v.s. permutation tests

Definitions:
- Population: The real-world data you are pulling from. Can have one or multiple distributions.
- Observation: A single datapoint drawn from a population (rows).
- Feature: A characteristic of an observation we are measuring (columns).
- Sample: A sub-population. The observations of a population that share a common feature or features.

*Note that some texts will use different words to describe this, or use these words to describe different things*

**(H)** I have a population distribution, and I have one sample. Does this sample look like it was drawn from the population?

**(P)** I have two samples, but no information about any population distributions. Do these samples look like they were drawn from the same population?
